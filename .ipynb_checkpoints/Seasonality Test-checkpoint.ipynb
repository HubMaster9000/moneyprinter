```python
import matplotlib.pyplot as plt
import pandas as pd
```

# Alpha Vantage 

```python
from alpha_vantage.timeseries import TimeSeries
ts = TimeSeries(key='JHNRDQM3HA7WWF08', output_format='pandas')
```

```python
data, meta_data = ts.get_intraday('GOOGL', interval='1min')
data = data.iloc[::-1]
```

```python
# Only for 11/25/2020
data_Filter = data.loc[pd.Timestamp('11/27/2020'):]
data_Filter.head()
```

```python
data_Filter['4. close'].plot()
plt.title('Intraday Times Series for the GOOGL stock (1 min)')
plt.show()
```

# Time Series Decomp

Get the log returns

```python
from statsmodels.tsa.seasonal import seasonal_decompose

result = seasonal_decompose(data['4. close'], model='additive', period=1)
result.plot()
plt.show()
```

```python
plt.plot(result.trend)
plt.plot(result.observed)
```
